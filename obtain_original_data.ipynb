```python
env = gym.make('hopper-medium-v2')
pickle_file_path = './step_1_train_scm_model/Original_datasets/hopper_v2/hopper-medium-v2.pickle'

dataset = env.get_dataset()
print(dataset['observations'].shape) 
print(dataset['actions'].shape) 
print(dataset['rewards'].shape) 
print(dataset['next_observations'].shape) 
print(dataset['terminals'].shape) 

newdataset = [dataset['observations'], dataset['actions'], dataset['rewards'], dataset['next_observations'], dataset['terminals']]

with open(pickle_file_path, 'wb') as f:
    pickle.dump(newdataset, f)
```
